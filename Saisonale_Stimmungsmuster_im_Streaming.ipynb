## 1.1 Forschungsfrage

> **Wie beeinflusst der Wechsel der Jahreszeiten unsere psychologische Stimmung und wie zeigt sich das in unseren Musik- und Videopräferenzen?**

Warum ist das gesellschaftlich relevant? Weil dahinter ein gut belegtes psychologisches Phänomen steht: die **saisonal-affektive Störung** (*„Winterdepression" oder "Winterblues"*). Mit kürzeren Tagen und weniger Licht sinkt bei vielen Menschen messbar die Stimmung. Der Körper produziert mehr Melatonin, der Serotoninhaushalt verschiebt sich. Stimmung ist also keine reine Privatsache, sondern folgt einem Rhythmus, den Millionen Menschen gleichzeitig erleben.

Das Spannende: Wir können diese Stimmung normalerweise nicht direkt messen. Aber jeden Tag treffen Hunderte Millionen Menschen eine kleine, ehrliche Entscheidung **welchen Song sie als Nächstes abspielen** oder **welche Serie sie als Nächstes gucken**. ## 1.2 Theoretische Grundlage: Audio-Features als Messgrundlage [1]

Bevor wir Millionen Songs analysieren können, brauchen wir eine Sprache, um Musik in Zahlen zu fassen. Spotify liefert für jeden Track sogenannte **Audio-Features**. Das sind Werte zwischen 0 und 1, die den Charakter eines Songs beschreiben. Diese vier (plus das Tempo) sind für uns zentral:

| Feature | Was es misst | 0.0 bedeutet … | 1.0 bedeutet … |
|---|---|---|---|
| **Valence** | musikalische *Positivität* / Stimmung | traurig, düster, niedergeschlagen | fröhlich, euphorisch, ausgelassen |
| **Energy** | Intensität & Aktivität | ruhig, sanft, zurückgelehnt | laut, schnell, druckvoll |
| **Danceability** | wie tanzbar (Rhythmus, Beat, Tempo) | schwer zu tanzen | „auf die Tanzfläche" |
| **Acousticness** | wie akustisch (unplugged vs. elektronisch) | elektronisch produziert | akustisch, unplugged |
| **Tempo** | Geschwindigkeit in BPM | – | – |

**Valence ist unser wichtigster Wert** er ist quasi das Stimmungsthermometer eines Songs. *Mad World* hat eine sehr niedrige Valence, *Vamos a la Playa* eine hohe. Wenn sich die Stimmung einer ganzen Bevölkerung verschiebt, sollte sich das im Durchschnitts-Valence der gehörten Musik zeigen. Die anderen drei Werte (Energy, Danceability, Acousticness) helfen uns später zu verstehen, *warum* die Stimmung kippt, denn sie sind die Bausteine, aus denen sich der emotionale Eindruck eines Songs zusammensetzt.

Um ein Gefühl für diese Werte zu bekommen, haben wir zunächst einige bekannte Songs anhand ihrer Features verglichen:



**Was man hier sieht:** Die beiden Eröffnungssongs liegen wie zwei Gegenpole an den Enden der Skala. *Mad World* fällt vor allem durch eine **sehr niedrige Valence** und hohe **Acousticness** auf. Es ist also eher ruhig und traurig. *Vamos a la Playa* dreht fast alle Regler ins Gegenteil: **hohe Valence, hohe Energy, hohe Danceability**, kaum akustisch. Die übrigen Songs reihen sich dazwischen ein und zeigen: Die Audio-Features bilden den intuitiven Eindruck eines Songs erstaunlich treffend ab. Genau deshalb können wir ihnen vertrauen, wenn wir gleich nicht mehr einzelne Songs, sondern ganze Länder und Jahreszeiten vermessen. [1]

## 1.3 Datensätze: Spotify und Netflix

Wir konzentrieren uns bewusst auf **zwei Datenquellen**, und das hat einen inhaltlichen Grund:

- **Spotify** verrät, *wie* wir uns fühlen. Musik läuft oft nebenbei und **automatisch**. Wir greifen zu dem, wonach uns *gerade* ist, ohne lange nachzudenken. Das macht das Hörverhalten zu einem ehrlichen Stimmungsbarometer.
- **Netflix** verrät, *was* wir mit dieser Stimmung anfangen. Abends vor dem Fernseher entscheiden wir uns dagegen **bewusst und gezielt** für ein bestimmtes Programm, sei es Action, Drama oder eine Komfort-Serie. Hier sehen wir Verhalten, nicht nur Gefühl.
- Andere Streamingdienste wie **Amazon Prime** oder **Disney Plus** haben wir bewusst ausgelassen. Netflix ist der klare Marktführer mit der breitesten Nutzerbasis weltweit. **Disney Plus** hingegen ist stark verzerrt, weil die Plattform fast nur Familien und Kinder anzieht. Da Kinder meist nur fröhliche und kindliche Filme in Dauerschleife schauen, hätte das unsere Analyse der echten gesellschaftlichen Stimmung komplett verfälscht.

Zusammen ergeben die beiden ein vollständigeres Bild: Das unbewusste Gefühl (Spotify) und die bewusste Reaktion darauf (Netflix).

**Die Datensätze im Überblick:**
- *Spotify Top-Charts* mit täglichen (2017-2023) / wöchentliche (2016-2022) Streams pro Land, jeweils angereichert um die Audio-Features (Valence, Energy, Danceability, Acousticness, Tempo) und der jeweilligen Region im wöchentlichen Datensatz. [1], [2], [3]


- *Netflix Top 10* mit den wöchentlich gestreamten Stunden pro Titel (2021 - heute), plus die dazugehörigen **Genre Informationen** aus den **Katalogen für Filme und Serien** (bis 2025). Wichtig ist hierbei, dass wir die Filme und Serien aus den Top 10 direkt mit den passenden Informationen zu ihrem Genre verknüpft haben, um die inhaltliche Richtung genau zu analysieren. [4], [5]

### Spotify Datensatz laden und bereinigen:
Alle Werte außerhalb des offiziellen Spotify Wertebereich entfernen und Header-Zeilen entfernen

````python
PFAD_SPOTIFY = 'Datensätze/final.csv'

def lade_spotify(pfad_csv: str) -> pd.DataFrame:
    print(f'Lese {pfad_csv} ...')
    df = pd.read_csv(pfad_csv, index_col=0)
    # Header-Zeilen entfernen, die sich während der Erhebung eingeschlichen haben
    df = df[df['country'] != 'country'].copy()
    df['week_date'] = pd.to_datetime(df['week'], errors='coerce')
    df = df.dropna(subset=['week_date'])
    # Numerische Audio-Features sicherstellen
    for col in ['valence', 'danceability', 'energy', 'acousticness',
                'tempo', 'loudness', 'streams', 'rank']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    # Ausreißer außerhalb der offiziellen Spotify-Wertebereiche als fehlend markieren
    df.loc[~df['valence'].between(0, 1),       'valence']      = float('nan')
    df.loc[~df['energy'].between(0, 1),        'energy']       = float('nan')
    df.loc[~df['danceability'].between(0, 1),  'danceability'] = float('nan')
    df.loc[~df['acousticness'].between(0, 1),  'acousticness'] = float('nan')
    df.loc[~df['loudness'].between(-60, 0),    'loudness']     = float('nan')
    df.loc[~df['tempo'].between(40, 250),      'tempo']        = float('nan')
    df['year']  = df['week_date'].dt.year
    df['month'] = df['week_date'].dt.month
    return df

def klassifiziere_saison(monat: int) -> str:
    if monat in (6, 7, 8):  return 'Sommer'
    if monat in (12, 1, 2): return 'Winter'
    return 'Übergang'

spotify = lade_spotify(PFAD_SPOTIFY)
spotify['saison'] = spotify['month'].apply(klassifiziere_saison)

print(f'Einträge gesamt:   {len(spotify):,}'.replace(',', '.'))
print(f'Zeitraum:         {spotify["week_date"].min():%Y-%m-%d} bis {spotify["week_date"].max():%Y-%m-%d}')
print(f'Länder erfasst:   {spotify["country"].nunique()}')
print(f'Regionen:         {spotify["region"].nunique()}')
spotify[['week_date','country','track_name','artist_genre','valence','energy','saison']].head()
````

# 2 Streaming-Rhythmus im Wochenverlauf [1], [2]

Bevor wir auf die große Bühne gehen, beginnen wir ganz klein mit einer einzigen Wochenbetrachtung. Verändert sich unsere Stimmung schon innerhalb von sieben Tagen? Um einen ersten Überblick zu bekommen, mitteln wir die vier wichtigsten Audio-Features für jeden Wochentag und erstellen daraus eine Heatmap um die Unterschiede im Wochenverlauf deutlich zu machen. So entsteht ein **„musikalischer Fingerabdruck"** der Woche.

````python
import matplotlib.pyplot as plt
import seaborn as sns

features_of_interest = ['Valence', 'Energy', 'Danceability', 'Acousticness']
# observed=False: alle 7 Wochentage behalten, auch wenn ein Tag wenig Einträge hat
daily_avg = df_daily.groupby('Weekday', observed=False)[features_of_interest].mean().reset_index()

tage_deutsch = {
    'Monday': 'Montag', 'Tuesday': 'Dienstag', 'Wednesday': 'Mittwoch',
    'Thursday': 'Donnerstag', 'Friday': 'Freitag', 'Saturday': 'Samstag', 'Sunday': 'Sonntag'
}
daily_avg['Weekday'] = daily_avg['Weekday'].map(tage_deutsch)

# .T dreht die Matrix: Features als Zeilen, Wochentage als Spalten (Woche liest man links → rechts)
heatmap_data = daily_avg.set_index('Weekday')[features_of_interest].T

# Z-Score pro Zeile (axis=1), damit jedes Feature auf seiner eigenen Skala normiert wird
heatmap_data_scaled = heatmap_data.apply(lambda x: (x - x.mean()) / x.std(), axis=1)

plt.figure(figsize=(12, 6))
# heatmap_data_scaled steuert die Farbe, heatmap_data liefert die echten Werte als Annotation
ax = sns.heatmap(
    heatmap_data_scaled,
    annot=heatmap_data,
    fmt=".3f",
    cmap="coolwarm",
    linewidths=1,
    cbar_kws={'label': 'Abweichung vom Wochendurchschnitt (Standardabweichungen)'}
)

cbar = ax.collections[0].colorbar
cbar.set_ticks([-1.0, 0.0, 1.0])
cbar.set_ticklabels(['−1σ (unterdurchschnittlich)', 'Ø Wochenmittel', '+1σ (überdurchschnittlich)'])

plt.title('Der musikalische Fingerabdruck der Woche', fontsize=18, pad=20, fontweight='bold')
plt.xlabel('')
plt.ylabel('Audio Features', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout()
plt.show()
````

### Was die Woche zeigt

Ein kleiner Methodenhinweis vorweg: Wir färben die Felder nicht nach dem *absoluten* Wert, sondern nach der **Abweichung vom Wochendurchschnitt** (Z-Score pro Feature). Das ist wichtig, weil die vier Features auf ganz unterschiedlichen Niveaus liegen. Die hohen Werte würden dann alles überstrahlen. Durch die Z-Score Standardisierung wird jede Zeile vergleichbar: Rot heißt „für diesen Wochentag überdurchschnittlich", Blau „unterdurchschnittlich". Die tatsächlichen Zahlen stehen weiterhin in den Feldern.

**Das auffälligste Muster ist die Spannung zwischen Wochenanfang und Wochenende.** Zum **Wochenende hin** wandern die stimmungs- und energiegeladenen Features (Valence, Energy, Danceability) nach oben die Musik wird hörbar fröhlicher, energiereicher und tanzbarer. Das passt perfekt zu unserer Alltagserfahrung: Freitag und Samstag wird gefeiert, die Playlist macht mit. **Acousticness** verhält sich dabei spiegelbildlich. Sie ist tendenziell zu Wochenbeginn höher, wenn ruhigere, akustischere Musik gefragt ist, und sinkt zum Wochenende, wenn elektronischere Party-Sounds übernehmen.

Die Kernbotschaft dieser ersten Visualisierung: **Unser Hörverhalten hat einen Rhythmus, und dieser Rhythmus spiegelt unsere Stimmung wider.** Wenn sich das schon über sieben Tage zeigt wie deutlich wird es dann erst über zwölf Monate?

### Aber: 
Wir haben keinen direkten Draht in die Köpfe der Menschen. Kein Datensatz sagt uns *„diese Person ist heute traurig"*. Aber genau hier liegt die Stärke der Audio-Features: Sie erlauben uns einen **Rückschluss**. Wenn in einem Zeitraum systematisch fröhlichere, energiereichere Musik gehört wird, ist es plausibel, dass die Menschen tendenziell besser gestimmt sind und umgekehrt.

Aus genau diesem Schluss leiten wir die Hypothese unserer Arbeit ab, die wir ab jetzt prüfen: 

**Im Winter sind die Menschen tendenziell trauriger, im Sommer fröhlicher.**

## 3.1 Methodischer Zwischenschritt: Warum wir Länder herausfiltern müssen

Die Weltkarte hat uns gerade selbst die Lösung verraten: Wenn Nord- und Südhalbkugel **gegenläufig** schwingen, dann heben sie sich gegenseitig auf, sobald wir alles in einen Topf werfen. # 4 Stimmungs-Barometer: Valence 2017–2022 [1], [3]

Jetzt kommt der Kern unserer Story. Wir tragen die durchschnittliche Valence der Nordhalbkugel über die zwölf Monate auf. Und zwar nicht nur als eine Linie, sondern mit einem **Schwankungsband**, das zeigt, wie verlässlich das Muster Jahr für Jahr wiederkehrt.

````python
monat_jahr_nh = spotify_nh.groupby(['year', 'month'])['valence'].mean().reset_index()

monat_stats_nh = (monat_jahr_nh.groupby('month')['valence']
                  .agg(mittel='mean', streuung='std', vmin='min', vmax='max')
                  .reset_index())

x_nh      = monat_stats_nh['month']
mittel_nh = monat_stats_nh['mittel']
streu_nh  = monat_stats_nh['streuung']

fig, ax = plt.subplots(figsize=(12, 6))

ax.axvspan(5.5,  8.5,  color=FARBE_SOMMER, alpha=0.12, zorder=0)
# Winter liegt an der Jahresgrenze → zwei separate Spans nötig (Dez + Jan/Feb)
ax.axvspan(11.5, 12.5, color=FARBE_WINTER, alpha=0.18, zorder=0)
ax.axvspan(0.5,  2.5,  color=FARBE_WINTER, alpha=0.18, zorder=0)

# Bänder von außen nach innen aufbauen (zorder 1→2→3), damit das engere Band das breitere überlagert
ax.fill_between(x_nh, monat_stats_nh['vmin'], monat_stats_nh['vmax'],
                color='#E2492C', alpha=0.12,
                label='Spannweite (Min–Max aller Jahre)', zorder=1)
ax.fill_between(x_nh, mittel_nh - streu_nh, mittel_nh + streu_nh,
                color='#E2492C', alpha=0.30,
                label='±1 Standardabweichung', zorder=2)
ax.plot(x_nh, mittel_nh, color='#E2492C', linewidth=3.2,
        marker='o', markersize=8, label='Ø alle Jahre', zorder=3)

from matplotlib.patches import Patch
legende_extra = [
    Patch(facecolor=FARBE_SOMMER, alpha=0.35, label='Sommer (Jun–Aug)'),
    Patch(facecolor=FARBE_WINTER, alpha=0.35, label='Winter (Dez–Feb)'),
]

ax.set_xticks(range(1, 13))
ax.set_xticklabels(MONATS_NAMEN)
ax.set_xlabel('Monat')
ax.set_ylabel('Durchschnittliche Valence (0 = traurig, 1 = fröhlich)')
ax.set_title('Stimmungs-Barometer: Valence mit Konfidenzband – Nordhalbkugel-Filter (2017–2022)')
ax.set_xlim(0.5, 12.5)

handles, labels = ax.get_legend_handles_labels()
# axvspan erzeugt keine Legende-Handles → Patch-Objekte manuell anhängen
ax.legend(handles=handles + legende_extra,
          loc='lower right', fontsize=9, framealpha=0.9)

plt.tight_layout()
plt.show()
````

### Was das Barometer verrät

**Die paradoxe Weihnachts-Spitze.** Das Auffälligste zuerst: Mitten im tiefsten Winter schießt die Valence im **Dezember** nach oben ausgerechnet im dunkelsten Monat. Das wirkt wie ein Widerspruch zu unserer These, ist aber leicht erklärt: **Weihnachtsmusik.** Klassiker wie *„Rocking Around the Christmas Tree"* erreichen Valence-Werte von oft über **0,90** und fluten im Dezember die Charts. Dieser kulturelle Sonderfall überlagert kurzzeitig die saisonale Grundstimmung. Es ist ein schönes Beispiel dafür, dass Daten immer im kulturellen Kontext gelesen werden müssen. Der Ausreißer ist kein Fehler, sondern eine Geschichte für sich.

**Klammert man Weihnachten aus, ist das Muster eindeutig.** Die Sommermonate (Jun–Aug, gold hinterlegt) liegen klar über den tiefen Wintermonaten Januar/Februar (blau hinterlegt). Genau das hatten wir vorhergesagt: hellere, fröhlichere Musik im Sommer, gedrücktere Stimmung im Spätwinter.

**Das Schwankungsband ist der eigentliche Beweis.** Das schmale ±1σ-Band zeigt, dass dieses Muster **über fünf Jahre hinweg (2017–2022) immer wieder fast identisch auftritt**. Es ist also kein Zufallsjahr, sondern ein stabiler, jährlich wiederkehrender Rhythmus.

**Ein Bonus-Fund am unteren Rand.** Schaut man auf die Spannweite, bildet das Jahr **2017** durchgehend die unterste Linie. Über die Jahre wandert die durchschnittliche Valence langsam nach oben. Wir hören also tendenziell von Jahr zu Jahr etwas fröhlichere Musik. Deswegen ist auch der niedrigste Wert der Spannweite genau das Jahr 2017, das das erste Jahr unseres Datensatzes ist.

Valence ist ein starker Einzelwert aber er fällt nicht vom Himmel. Er lässt sich als **Zusammenspiel der anderen drei Features** verstehen: Energy, Danceability und Acousticness formen gemeinsam den emotionalen Eindruck, den wir als „fröhlich" oder „traurig" wahrnehmen. Schauen wir uns also an, *welche* dieser Bausteine die saisonale Welle eigentlich antreiben.

# 6 Genre-Präferenzen: Was schauen wir wann? [4], [5]

Wir wechseln die Bühne. Für jedes Netflix-Genre berechnen wir, wie groß sein Anteil am Sommer- bzw. Winter-Konsum ist gewichtet nach gestreamten Stunden, damit populäre Titel stärker zählen. Die Balken zeigen die **Differenz**: Wie viel öfter (in Prozentpunkten) wird ein Genre im Sommer relativ zum Winter geschaut?

````python
# Genre-Lookup aus Filmen UND Serien zusammenführen
_lookup_filme  = netflix_filme[['title', 'genres']].dropna(subset=['genres']).drop_duplicates('title')
_lookup_serien = netflix_serien[['title', 'genres']].dropna(subset=['genres']).drop_duplicates('title')
nf_lookup = pd.concat([_lookup_filme, _lookup_serien]).drop_duplicates('title')
nf_lookup['title_norm'] = nf_lookup['title'].str.lower().str.strip()

nf = netflix_top.copy()
nf['title_norm'] = nf['show_title'].str.lower().str.strip()
nf = nf.merge(nf_lookup[['title_norm', 'genres']], on='title_norm', how='left')

# Jede Zeile in Einzel-Genres aufsplitten
nf_genres = nf.dropna(subset=['genres']).copy()
nf_genres['genres_list'] = nf_genres['genres'].str.split(', ')
nf_long = nf_genres.explode('genres_list')

# Gewichten mit den weekly_hours_viewed, damit populärere Titel mehr zählen
nf_long['gewicht'] = nf_long['weekly_hours_viewed'].fillna(1)

saison_pivot = (nf_long[nf_long['saison'] != 'Übergang']
                .groupby(['saison', 'genres_list'])['gewicht'].sum()
                .unstack('saison').fillna(0))

saison_anteil = saison_pivot.div(saison_pivot.sum(axis=0), axis=1) * 100

# Differenz Sommer minus Winter: positiv = sommerlastig, negativ = winterlastig
saison_anteil['diff'] = saison_anteil['Sommer'] - saison_anteil['Winter']
saison_anteil = saison_anteil.sort_values('diff')

fig, ax = plt.subplots(figsize=(10, 7))
farben = ['#3D7EAE' if d < 0 else '#FFB400' for d in saison_anteil['diff']]
ax.barh(saison_anteil.index, saison_anteil['diff'], color=farben, edgecolor='white')
ax.axvline(0, color='#555', lw=1)
ax.set_xlabel('Sommer-Anteil minus Winter-Anteil (Prozentpunkte)')
ax.set_title('Welche Netflix-Genres werden im Sommer / Winter relativ öfter geschaut?')
ax.text(0.98, 0.02, 'positiv = sommerlastig\nnegativ = winterlastig',
        transform=ax.transAxes, ha='right', va='bottom', fontsize=9,
        bbox=dict(facecolor='white', alpha=0.7, edgecolor='#ccc'))
plt.tight_layout()
plt.show()
````

### Was unsere Genre-Wahl verrät

**Action ist Sommer, gemütlich ist Winter.** Die goldenen Balken (sommerlastig) werden von energiegeladenen Genres wie **Action** angeführt das passt nahtlos zur höheren Sommer-Energy bei Spotify: Wenn wir aktiv und aufgedreht sind, wollen wir auch Action sehen. Die blauen Balken (winterlastig) gehören dagegen den **ruhigeren, gehaltvolleren Genres wie Crime und Drama**. Bildlich gesprochen: Im Winter kuschelt man sich nicht mit einem Action-Kracher unter die Decke, sondern mit etwas Ruhigem, das einen in die Geschichte hineinzieht. Die bewusste Netflix-Wahl folgt damit demselben Stimmungsmuster wie die unbewusste Spotify-Wahl ein schönes Stück gegenseitiger Bestätigung über zwei völlig verschiedene Datensätze hinweg.

**Kids-Inhalte bleiben konstant die Kontrollgruppe.** Kinder-Genres zeigen kaum saisonale Verschiebung, und das ist methodisch wertvoll: Was Kinder schauen, hängt nicht von der Stimmung Erwachsener ab, sondern von einer festen Alterskategorie. Dass dieser Balken nahe null bleibt, wirkt fast wie eine eingebaute Kontrollgruppe sie zeigt, dass die Ausschläge bei Action, Crime und Drama echte Stimmungseffekte sind und nicht bloß Datenrauschen.

Bisher ging es darum, *was* wir schauen. Es gibt aber noch eine einfachere, fast traurige Vermutung: Wer niedergeschlagen ist, zieht sich eher zurück und ist mehr allein und allein heißt oft: mehr Netflix. Wenn das stimmt, müsste sich nicht nur das *Genre*, sondern auch die schiere *Menge* des Konsums saisonal verschieben. Schauen wir nach.

# Netflix: Wie viel schauen wir im Jahresverlauf? [4], [5]

Zum Abschluss messen wir die reine Streaming-*Menge*. Pro Monat mitteln wir die wöchentlich gestreamten Gesamtstunden ein fairer Vergleich, unabhängig davon, wie oft ein Kalendermonat im Datensatz vorkommt.

````python
# Erst Gesamtstunden pro Woche summieren, dann pro Monat mitteln —
# direkt über Monate mitteln wäre falsch, da manche Monate mehr Wochen im Datensatz haben
wochen_summe = (
    netflix_top
    .groupby('week')['weekly_hours_viewed']
    .sum()
    .reset_index()
)
wochen_summe['month'] = wochen_summe['week'].dt.month

# Wie viele Wochen pro Monat sind im Datensatz? (Transparenz-Check)
wochen_pro_monat = wochen_summe.groupby('month')['week'].count().rename('n_wochen')

vol_monat = (
    wochen_summe
    .groupby('month')['weekly_hours_viewed']
    .mean()
    .reset_index()
)
vol_monat['stunden_mrd'] = vol_monat['weekly_hours_viewed'] / 1e9
vol_monat = vol_monat.merge(wochen_pro_monat, on='month')

MONATS_NAMEN = ['Jan', 'Feb', 'Mär', 'Apr', 'Mai', 'Jun', 'Jul', 'Aug', 'Sep', 'Okt', 'Nov', 'Dez']
vol_monat['monat_name'] = vol_monat['month'].apply(lambda m: MONATS_NAMEN[m-1])

def saison_farbe(m):
    if m in (6, 7, 8):   return FARBE_SOMMER
    if m in (12, 1, 2):  return FARBE_WINTER
    return '#B5B5B5'

farben = vol_monat['month'].apply(saison_farbe)

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(vol_monat['monat_name'], vol_monat['stunden_mrd'],
              color=farben, edgecolor='white', linewidth=1.2, width=0.7)

# Wert + Stichprobengröße über jeden Balken schreiben
for bar, val, n in zip(bars, vol_monat['stunden_mrd'], vol_monat['n_wochen']):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.03,
            f'{val:.3f}\n(n={n})', ha='center', va='bottom', fontsize=9)

from matplotlib.patches import Patch
ax.legend(handles=[
    Patch(color=FARBE_SOMMER, label='Sommer (Jun–Aug)'),
    Patch(color=FARBE_WINTER, label='Winter (Dez–Feb)'),
    Patch(color='#B5B5B5',   label='Übergang'),
], fontsize=10, loc='upper center', bbox_to_anchor=(0.5, -0.12),
          ncol=3, frameon=False)

ax.set_xlabel('Monat')
ax.set_ylabel('Ø Streaming-Stunden pro Woche (Mrd.)')
ax.set_title('Netflix-Streaming-Volumen: Ø wöchentliche Stunden pro Monat', fontsize=14, pad=12)
sns.despine()
plt.tight_layout()
plt.show()
````

### Was die Menge verrät

Das Bild schließt den Kreis: Die **Wintermonate (blau) ragen über die Sommermonate (gold) hinaus** im Winter wird pro Woche spürbar mehr gestreamt. Die kleinen `n=`-Angaben über den Balken sorgen dabei für Ehrlichkeit: Sie zeigen, auf wie vielen Wochen jeder Monatswert beruht, sodass man den Vergleich richtig einordnen kann.

Inhaltlich fügt sich das nahtlos in unsere Story: Wenn die Tage kurz und kalt sind, die Stimmung gedrückt ist und man ohnehin mehr Zeit drinnen und allein verbringt, dann wächst der Netflix-Konsum. **Mehr Netflix bedeutet oft mehr Zeit allein und mehr Zeit allein verstärkt tendenziell die gedrückte Winterstimmung.** Hier schließt sich ein Kreis: Dieselbe Jahreszeit, die uns bei Spotify zu ruhigerer, akustischerer Musik greifen lässt, treibt uns abends in größerer Menge vor den Bildschirm.

## 8.1 Wochenverlauf-Heatmap
[→ Zum Code](#abschnitt-1)

**Was wird berechnet:** Pro Wochentag mitteln wir die vier Audio-Features und berechnen daraus den **Z-Score** also wie weit jeder Wert vom Wochendurchschnitt abweicht. Das ist nötig, weil die Features auf völlig unterschiedlichen Niveaus liegen: Valence ~0.60, Acousticness ~0.25. Ohne Normalisierung würde die Farbe nur das Feature-Niveau zeigen, nicht die relative Schwankung über die Woche.

Der Z-Score normiert jedes Feature auf *seine eigene* Skala:

$$z = \frac{x - \mu}{\sigma}$$

- `x` – Tagesmittelwert des Features (z. B. Valence am Montag)
- `μ` – Wochendurchschnitt dieses Features
- `σ` – Standardabweichung über die Woche

**Beispiel:** Hat Valence am Montag den Wert 0.591, Wochenmittel μ = 0.597 und σ = 0.005, ergibt sich `z = (0.591 − 0.597) / 0.005 = −1.2`. Das Feld erscheint dunkelblau obwohl der absolute Unterschied nur 0.006 beträgt, macht der Z-Score ihn farblich sofort sichtbar. Ein positiver Z-Score (Wert über dem Schnitt) erscheint rot, ein negativer (darunter) blau; der Wochenschnitt selbst liegt immer bei 0.

Die Normierung passiert pro Feature (Zeile), da `axis=1` gesetzt ist so sind alle vier Features auf der Farbskala direkt vergleichbar, unabhängig von ihrem absoluten Niveau.

**Warum Heatmap:** Wir haben zwei kategoriale Achsen (Feature × Wochentag) und einen numerischen Farbwert das ist das klassische Einsatzgebiet einer Heatmap.

**Code-Entscheidungen:**

`sns.heatmap` bekommt zwei verschiedene DataFrames: `heatmap_data_scaled` (Z-Score) steuert die Farbe, `heatmap_data` liefert die echten Werte als Annotation. So sieht der Leser immer den tatsächlichen Spotify-Wert, aber die Farbe zeigt die relative Abweichung.

```python
ax = sns.heatmap(heatmap_data_scaled, annot=heatmap_data, fmt=".3f", ...)
```

## 8.3 Valence-Barometer mit Schwankungsband
[→ Zum Code](#abschnitt-3)

**Was berechnet:** Pro Monat berechnen wir Mittelwert, Standardabweichung sowie Minimum und Maximum der Valence über alle Jahre (2017–2022). Das ergibt drei Bänder: Min–Max zeigt den Extrembereich, ±1σ den normalen Schwankungsbereich, die Linie den Durchschnitt. Ein schmales Band bedeutet, alle Jahre zeigen dasselbe Muster ein breites Band weist auf Ausreißerjahre hin.

**Warum Linienchart mit Band:** Eine Zeitreihe mit Unsicherheit lässt sich so kompakter darstellen als mit Fehlerbalken, und man sieht sofort wie stabil der Effekt über die Jahre ist.

**Code-Entscheidungen:**

Der Winter umfasst Dezember, Januar und Februar aber die x-Achse ist linear (1–12), nicht ringförmig. Deshalb braucht es zwei separate `axvspan`-Aufrufe:

```python
ax.axvspan(11.5, 12.5, ...)  # Dezember
ax.axvspan(0.5,  2.5,  ...)  # Januar + Februar
```

Die zwei `fill_between`-Schichten bauen sich von außen nach innen auf (`zorder=1` → `zorder=2` → `zorder=3`), damit das engere σ-Band das breitere Min–Max-Band überlagert und nicht umgekehrt. Da `axvspan` keine Legende-Handles erzeugt, werden manuell `Patch`-Objekte erstellt und an die bestehende Legende angehängt.

## 8.4 Audio-Features im 2×2-Grid
[→ Zum Code](#abschnitt-4)

**Was berechnet:** Monatlicher Durchschnitt für Energy, Danceability, Acousticness und Tempo jeweils nur für Nordhalbkugel-Länder. Zusätzlich berechnen wir das Sommer- (Jun–Aug) und Winter-Mittel (Dez–Feb) als gestrichelte Referenzlinie pro Feature.

**Warum 2×2-Grid:** Vier Features auf einem einzigen Liniendiagramm wären schwer lesbar. Das Grid erlaubt jedem Feature eine eigene Y-Achse und damit eine optimale Skalierung.

**Code-Entscheidungen:**

`sharey=False` ist hier zwingend: Energy liegt um 0.65, Acousticness um 0.30, Tempo bei ~120 BPM eine gemeinsame Y-Achse würde die kleineren Features auf eine flache Linie zusammenquetschen.

```python
ax.fill_between(y.index, y, y.min() * 0.998, ...)
```

Statt bis 0 zu füllen, füllen wir nur bis knapp unter den niedrigsten Wert. Das hält die Y-Achse nah an den Daten und macht die Schwankungen sichtbar bei `fill_between(..., 0)` würde z. B. Acousticness auf einer Achse von 0 bis 0.32 kaum zu erkennen sein.

`MultipleLocator` setzt saubere Tick-Abstände: 0.01 für normalisierte Features (0–1), 1.0 für Tempo (BPM) sonst wählt Matplotlib unrunde Abstände.

## 8.6 Netflix-Streaming-Volumen nach Monat
[→ Zum Code](#abschnitt-6)

**Was berechnet:** Wir messen das reine Streaming-Volumen (Stunden) pro Monat. Die Aggregation passiert in zwei Schritten: erst alle Einträge einer Woche summieren, dann die Wochenwerte pro Monat mitteln. Der zweite Schritt ist wichtig direkt über Monate zu mitteln wäre falsch, weil manche Monate mehr Wochen im Datensatz haben als andere.

**Warum Balkendiagramm:** Absolute Mengen vergleichen ist die Kernaufgabe von Balkendiagrammen. Die Einfärbung nach Saison (Gold/Blau/Grau) macht den saisonalen Effekt sofort sichtbar.

**Code-Entscheidungen:**

```python
wochen_summe = netflix_top.groupby('week')['weekly_hours_viewed'].sum()
vol_monat    = wochen_summe.groupby('month')['weekly_hours_viewed'].mean()
```

Die Rohdaten liegen im Milliardenbereich. Die Division durch `1e9` macht die Y-Achse lesbar (`2.4` statt `2.400.000.000`).

Über jedem Balken steht `(n=X)` die Anzahl der Wochen, auf denen der Wert basiert. Das schafft Transparenz: Ein Monat mit nur 3 Wochen Datenbasis ist weniger belastbar als einer mit 8, und das sollte der Leser wissen.

# 9. Fazit

Wir sind mit einer simplen Frage gestartet: *Würdest du „Mad World" eher im Winter und „Vamos a la Playa" eher im Sommer hören?* Aus diesem Bauchgefühl ist über sechs Visualisierungen ein erstaunlich kohärentes Bild geworden.

**Unsere Forschungsfrage lautete:** *Wie beeinflusst der Wechsel der Jahreszeiten unsere psychologische Stimmung und wie zeigt sich das in unseren Musik- und Videopräferenzen?*

**Die Antwort der Daten ist deutlich und sie erzählt sich in einer durchgehenden Kette:**

1. Schon **innerhalb einer Woche** schwankt unser Hörverhalten rhythmisch mit der Stimmung (fröhlicher zum Wochenende).
2. **Weltweit** pulsieren Nord- und Südhalbkugel gegenläufig der überzeugendste Beleg dafür, dass die *Jahreszeit* der Taktgeber ist, nicht der Zufall.
3. Auf der Nordhalbkugel folgt die **Valence** einer stabilen, über fünf Jahre wiederkehrenden Jahreswelle (mit der kulturellen Weihnachts-Ausnahme im Dezember).
4. Angetrieben wird diese Welle vor allem von **Energy und Acousticness** im Sommer energiegeladen, im Winter ruhig und akustisch.
5. Bei der **bewussten** Netflix-Wahl zeigt sich dasselbe Muster: Action im Sommer, ruhiges Drama/Crime im Winter.
6. Und im Winter steigt sogar die **Gesamtmenge** des Konsums mehr Zeit allein, mehr Bildschirm.

Das Bemerkenswerte ist, dass **zwei völlig unabhängige Datensätze Spotify und Netflix dieselbe Geschichte erzählen.** Das unbewusste Gefühl (welcher Song läuft nebenbei) und die bewusste Entscheidung (was wähle ich heute Abend) zeigen in dieselbe Richtung. Genau diese Übereinstimmung macht das Ergebnis robust.

### Warum ist das wichtig?
Hinter den bunten Diagrammen steht ein reales Gesundheitsthema: die saisonal bedingte Stimmungstief. Unsere Daten zeigen, dass dieses Phänomen kein subjektives Einzelgefühl ist, sondern sich im kollektiven Verhalten von Millionen Menschen messbar niederschlägt. Streaming-Daten könnten so zu einem **frühen, anonymen Stimmungsbarometer einer Gesellschaft** werden.

### Kritische Einordnung
Wir sollten ehrlich bleiben: **Korrelation ist nicht Kausalität.** Wir messen nicht direkt die Stimmung der Menschen, sondern schließen sie aus Musik- und Videoverhalten ein indirekter, plausibler, aber nicht beweisender Weg. Faktoren wie Schulferien, Feiertage, Wetter oder Marketing-Kampagnen spielen ebenfalls mit hinein. Die Audio-Features sind außerdem algorithmisch geschätzt und nicht perfekt. Unsere Story zeigt ein starkes, konsistentes *Muster* sie liefert keine medizinische Diagnose.
